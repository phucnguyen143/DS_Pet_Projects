### Quick web crawl and sentiment analysis exercise

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
from textblob import TextBlob
```

```python
url = 'https://twitter.com/realDonaldTrump/status/984022625440747520'
source_doc = requests.get(url).text
source_html = BeautifulSoup(source_doc, 'html.parser')
```

```python
source_html
```

```python
type(source_html)
```

```python
twt_area = source_html.find(id = 'stream-items-id')
twt_area
```

```python
tweets = twt_area.findAll('div')
tweet = [tweet for tweet in tweets if ''.join(tweet.get('class')) == 'js-tweet-text-container']
tweet_post = [twt.get_text() for twt in tweet]
tweet_post
```

```python
result = []

for post in tweet_post:
    score = TextBlob(post)
    result.append(score.sentiment.polarity)

print(result)
```

```python
result_np = np.array(result)
```

```python
result_np
```

```python
result_np.mean()
```
